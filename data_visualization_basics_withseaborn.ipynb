### Creating Visualizations with the data

```python
seaborn.pairplot(iris, hue='species', height=3, aspect=1) # visualizes the distribution
```

```python
iris.hist(edgecolor='black', linewidth=1.2, figsize=(12, 8)) # histograms
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
seaborn.violinplot(x='species', y='sepal_length', data=iris)
plt.subplot(2, 2, 2)
seaborn.violinplot(x='species', y='sepal_length', data=iris)
plt.subplot(2, 2, 3)
seaborn.violinplot(x='species', y='sepal_length', data=iris)
plt.subplot(2, 2, 4)
seaborn.violinplot(x='species', y='sepal_length', data=iris)
```

```python
iris.boxplot(by='species', figsize=(12, 8))
```
